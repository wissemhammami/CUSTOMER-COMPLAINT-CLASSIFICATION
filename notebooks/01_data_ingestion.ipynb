```python
import pandas as pd

# 1. Load the raw CSV
df = pd.read_csv('data/raw/complaints_raw.csv')

# 2. Keep only the two columns you need
df = df[['Consumer complaint narrative', 'Product']]

# 3. Drop rows where complaint narrative is missing
df = df.dropna(subset=['Consumer complaint narrative'])

# 4. Rename columns to snake_case for easier coding
df = df.rename(columns={
    'Consumer complaint narrative': 'text',
    'Product': 'label'
})

# 5. Quick look
print(df.shape)
print(df['label'].value_counts())
print(df.head())

# 6. Save to processed folder
df.to_csv('data/processed/complaints_clean.csv', index=False)

# Check no label is missing after filtering
print(df['label'].isna().sum())

# Confirm class distribution as percentages
print(df['label'].value_counts(normalize=True).round(3) * 100)
```
