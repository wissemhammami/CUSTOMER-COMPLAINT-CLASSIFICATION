```python
import pandas as pd

# Load cleaned data
df = pd.read_csv('data/processed/complaints_clean.csv')

# 1. Shape
print('Shape:', df.shape)

# 2. Column types
print(df.dtypes)

# 3. Missing values
print('Missing values:')
print(df.isna().sum())

# 4. Duplicate rows
print('Duplicates:', df.duplicated().sum())

# 5. Label distribution
print('Label counts:')
print(df['label'].value_counts())

# 6. Text length stats
df['text_length'] = df['text'].apply(len)
print('Text length stats:')
print(df['text_length'].describe())

# 7. Check for empty strings
empty = (df['text'].str.strip() == '').sum()
print('Empty text strings:', empty)
```

```python
# 1. Remove duplicates
df = df.drop_duplicates()
print('After removing duplicates:', df.shape)

# 2. Remove very short texts (less than 50 characters)
df = df[df['text'].str.len() >= 50]
print('After removing short texts:', df.shape)

# 3. Reset index
df = df.reset_index(drop=True)

# 4. Final label distribution
print(df['label'].value_counts())

# 5. Save validated data
df.to_csv('data/processed/complaints_validated.csv', index=False)
print('saved successfully')
```
