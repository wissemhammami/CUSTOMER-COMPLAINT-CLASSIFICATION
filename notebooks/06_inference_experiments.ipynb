```python
import pandas as pd
import joblib
import numpy as np

# Load champion model
model = joblib.load('models/latest/model.pkl')

# Test with sample complaints
sample_complaints = [
    "I have been charged twice for the same transaction on my checking account and the bank refuses to refund me.",
    "My mortgage payment was incorrectly applied and I am now being reported as late to the credit bureau.",
    "I sent money through a wire transfer service and the money never arrived at the destination.",
    "My student loan servicer keeps changing and I have lost track of my payments and balance.",
    "The dealership added extra fees to my auto loan without my consent after I signed the contract."
]

# Predict
predictions = model.predict(sample_complaints)

# Confidence scores
proba = model.decision_function(sample_complaints)
confidence = np.max(proba, axis=1)

# Display results
results = pd.DataFrame({
    'complaint': sample_complaints,
    'predicted_label': predictions,
    'confidence_score': confidence.round(3)
})

pd.set_option('display.max_colwidth', 60)
print(results.to_string(index=False))
```
